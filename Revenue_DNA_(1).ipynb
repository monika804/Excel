# **Univariate Numerical Analysis**

## 1. HistPlot

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

df = pd.read_csv('/content/Revenue_DNA_Complete_Dataset_10000_Rows.csv')
sns.histplot(df["Net_Revenue"], bins=30)
plt.title("Net Revenue Distribution")
plt.show()
```

# 2. BoxPlot

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

df = pd.read_csv('/content/Revenue_DNA_Complete_Dataset_10000_Rows.csv')
plt.figure()
sns.boxplot(x=df["Net_Revenue"])
plt.title("Net_Revenue Boxplot")
plt.show()
```

# 3. Distplot

```python
plt.figure()
sns.histplot(
    df["Net_Revenue"],
    bins=30,
    kde=True,
    stat="density",
    alpha=0.6
)

plt.title("Distribution of Net Revenue with KDE Curve", fontsize=14, fontweight="bold")
plt.xlabel("Net_Revenue")
plt.ylabel("Density")
plt.tight_layout()
plt.show()
```

# 4. KDE Plot

```python
plt.figure()
sns.kdeplot(
    data=df,
    x="Net_Revenue",
    fill=True,
    linewidth=2,
    alpha=0.6
)

plt.title("KDE Plot of Revenue Distribution", fontsize=14, fontweight="bold")
plt.xlabel("Net_Revenue")
plt.ylabel("Density")
plt.tight_layout()
plt.show()
```

# 5. Violin plot

```python
plt.figure()
sns.violinplot(
    y=df["Net_Revenue"],
    inner="quartile"
)

plt.title("Violin Plot of Revenue", fontsize=14, fontweight="bold")
plt.ylabel("Net_Revenue")
plt.tight_layout()
plt.show()
```

# **Univariate Categorical**

# 1. Bar plot

```python
import matplotlib.pyplot as plt

# Data (you can adjust values based on your dataset)
regions = ['Canada', 'UK', 'UAE', 'India', 'USA']
revenue = [18200000, 17800000, 17600000, 17200000, 16900000]

plt.figure(figsize=(8,5))
plt.bar(regions, revenue)

plt.title('Revenue by Region')
plt.xlabel('Region')
plt.ylabel('Net Revenue')

plt.show()
```

# 2.  Pie chart

```python
import matplotlib.pyplot as plt


category_revenue = df.groupby("Product_Plan")["Net_Revenue"].sum()

plt.figure(figsize=(7,7))
plt.pie(
    category_revenue,
    labels=category_revenue.index,
    autopct="%1.1f%%",
    startangle=140,
    wedgeprops={"edgecolor": "black"}
)

plt.title("Revenue Contribution by Product", fontsize=14, fontweight="bold")
plt.tight_layout()
plt.show()
```

# ** **Revenue Drivers Analysis** **

# Correlation Analysis

```python
corr_matrix = df.corr(numeric_only=True)
corr_matrix["Net_Revenue"].sort_values(ascending=False)
```

# Visual Correlation Heatmap

```python
plt.figure(figsize=(10,8))
sns.heatmap(corr_matrix, cmap="coolwarm")
plt.title("Correlation Heatmap")
plt.show()
```
